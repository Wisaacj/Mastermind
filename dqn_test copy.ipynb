# Training a DQN agent to play Mastermind

## Imports & Constants

```python
from agents import MlpDQNAgent
from environment import Mastermind, Code
```

## Training

```python
env = Mastermind(num_colours=4, code_length=2)
```

```python
seed = 777
num_frames = 10_000
memory_size = 1_000
gamma = 0.99
batch_size = 32
target_update = 50
epsilon_decay = 1 / (num_frames // 2)

agent = MlpDQNAgent(
    env,
    memory_size,
    batch_size,
    target_update,
    epsilon_decay,
    seed,
    gamma,
)
```

```python
scores, losses, epsilons = agent.train(num_frames)
```

```python
agent.test(10, time_interval=1)
```

# Training a DQN agent to play Mastermind

## Imports & Constants

```python
import gymnasium as gym

from dqn.agents import MlpDQNAgent, MlpNoisyDQNAgent, RainbowDQNAgent, MlpDQNWithPERAgent
from environment import Mastermind, Code
```

## Training

```python
env = Mastermind(num_colours=3, code_length=3)
```

```python
seed = 777
num_frames = 200_000
memory_size = 100_000
gamma = 0.99
batch_size = 256
target_update = 2_000
epsilon_decay = 1 / (num_frames // 10)

agent = MlpDQNWithPERAgent(
    env=env,
    memory_size=memory_size,
    batch_size=batch_size,
    target_update=target_update,
    epsilon_decay=epsilon_decay,
    seed=seed,
    gamma=gamma,
    min_epsilon=0.05,
)
```

```python
scores, losses, epsilons = agent.train(num_frames)
```

```python
agent.env = env
agent.test(10, time_interval=0)
```
